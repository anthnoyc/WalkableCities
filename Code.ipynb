```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from scipy.optimize import curve_fit

# Read the data from CSV files
My_Data = pd.read_csv("/Users/dayfamily/Desktop/Projects/DataRes_Spring_2023/walkable_vehicles.csv")
My_Data_Original = pd.read_csv("/Users/dayfamily/Desktop/Projects/DataRes_Spring_2023/walkable_score.csv")

# Merge the data based on the 'City' column
My_Joint_Data = pd.merge(My_Data_Original, My_Data, on="City")

# Extract the x and y data
x_data = np.array(My_Joint_Data['Walk Score'])
y_data = np.array(My_Joint_Data['# of Vehicles per Household, 2019'])

# Define the model function
def model_f(x, a, b, c):
    return a * x + b * x**2 + c

# Fit the curve to the data
popt, pcov = curve_fit(model_f, x_data, y_data)
a_opt, b_opt, c_opt = popt

# Generate y values using the fitted model
x_model = np.linspace(min(x_data), max(x_data), 100)
y_model = model_f(x_model, a_opt, b_opt, c_opt)

# Plot the data and the fitted curve
plt.figure(figsize=(8, 6))
plt.xlabel("Walk Score")
plt.ylabel("Number of Vehicles Per Household")
plt.title("Walk Score vs. Number of Vehicles Per Household")
plt.grid(True)
plt.scatter(x_data, y_data, color='blue', s=50, alpha=0.7)
plt.plot(x_data, y_model, color='orange')
plt.legend(['Fitted Curve', 'Data Points'])
plt.xticks(np.arange(min(x_data), max(x_data)+1, 10))
plt.show()


```

```python
def get_color_gradient(c2, c1, n):
    """
    Given two hex colors, returns a color gradient
    with n colors.
    """
    assert n > 1
    c1_rgb = np.array(hex_to_RGB(c1))/255
    c2_rgb = np.array(hex_to_RGB(c2))/255
    mix_pcts = [x/(n-1) for x in range(n)]
    rgb_colors = [((1-mix)*c1_rgb + (mix*c2_rgb)) for mix in mix_pcts]
    return ["#" + "".join([format(int(round(val*255)), "02x") for val in item]) for item in rgb_colors]
```
